```python
import pandas as pd
import numpy as np
from google.colab import files
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.vector_ar.vecm import coint_johansen
import matplotlib.pyplot as plt

# Lista de pares a analizar
pairs = [("XOM", "JNJ"), ("JNJ", "PG"), ("XOM", "PG")]

# Función para cargar datos
def load_data(ticker):
    print(f"Por favor, sube el archivo Excel para {ticker}")
    upload = files.upload()
    filename = list(upload.keys())[0]
    df = pd.read_excel(filename)
    return df

# Función para preparar datos
def prepare_data(df1, df2, ticker1, ticker2):
    print(f"\nColumnas del DataFrame {ticker1}:", list(df1.columns))
    print(f"Columnas del DataFrame {ticker2}:", list(df2.columns))

    columna_esperada = " Close"
    if columna_esperada in df1.columns and columna_esperada in df2.columns:
        columna_cierre = columna_esperada
        print(f"Se encontró '{columna_esperada}' en ambos DataFrames, usándola para precios de cierre")
    else:
        while True:
            columna_cierre = input(f"Ingresa la columna de precios de cierre para {ticker1} y {ticker2}: ").strip()
            if columna_cierre in df1.columns and columna_cierre in df2.columns:
                break
            else:
                print(f"Error: '{columna_cierre}' no encontrada en uno o ambos DataFrames")

    close1 = df1[columna_cierre].dropna()
    close2 = df2[columna_cierre].dropna()
    min_len = min(len(close1), len(close2))
    return close1[:min_len], close2[:min_len]

# Pruebas de raíz unitaria (ADF y KPSS)
def unit_root_tests(series, name):
    print(f"\nPruebas de raíz unitaria para {name}:")
    adf_result = adfuller(series)
    print("Prueba ADF:")
    print(f'Estadístico ADF: {adf_result[0]:.4f}, p-valor: {adf_result[1]:.4f}')
    print("Interpretación:", "Estacionario" if adf_result[1] < 0.05 else "No estacionario (posible raíz unitaria)")

    kpss_result = kpss(series)
    print("\nPrueba KPSS:")
    print(f'Estadístico KPSS: {kpss_result[0]:.4f}, p-valor: {kpss_result[1]:.4f}')
    print("Interpretación:", "No estacionario" if kpss_result[1] < 0.05 else "Estacionario")

# Prueba de cointegración
def cointegration_test(close1, close2, ticker1, ticker2):
    df = pd.DataFrame({ticker1: close1, ticker2: close2})
    result = coint_johansen(df, det_order=0, k_ar_diff=1)
    print(f"\nPrueba de cointegración Johansen para {ticker1} y {ticker2}:")
    print(f"Estadístico de traza: {result.lr1}")
    print(f"Valores críticos (95%): {result.cvt[:, 1]}")
    if result.lr1[0] > result.cvt[0, 1]:
        print(f"Conclusión: {ticker1} y {ticker2} están cointegrados (relación a largo plazo)")
    else:
        print(f"Conclusión: No hay cointegración entre {ticker1} y {ticker2}")

# Modelo ARMA
def find_best_arma(series, name, max_p=3, max_q=3):
    mejor_aic = float('inf')
    mejor_orden = None
    for p in range(max_p + 1):
        for q in range(max_q + 1):
            try:
                modelo = ARIMA(series, order=(p, 0, q))
                resultados = modelo.fit()
                if resultados.aic < mejor_aic:
                    mejor_aic = resultados.aic
                    mejor_orden = (p, 0, q)
            except:
                continue
    if mejor_orden is None:
        mejor_orden = (1, 0, 1)
    print(f"Mejor modelo ARMA para {name}: {mejor_orden}, AIC: {mejor_aic:.2f}")
    return mejor_orden

# Graficar pronósticos con interpretación
def plot_forecast(model, series, name, steps=30):
    forecast_obj = model.get_forecast(steps=steps)
    pronostico = forecast_obj.predicted_mean
    conf_int = forecast_obj.conf_int()
    indice = range(len(series), len(series) + steps)

    plt.figure(figsize=(12, 6))
    plt.plot(series, label=f'Histórico {name}')
    plt.plot(indice, pronostico, label='Pronóstico', color='red')
    plt.fill_between(indice, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='pink', alpha=0.3, label='95% CI')
    plt.title(f'Pronóstico ARMA de {name}')
    plt.legend()
    plt.show()

    ultimo = series.iloc[-1]
    prom_pronostico = pronostico.mean()
    print(f"\n{name} - Último valor: {ultimo:.2f}, Pronóstico promedio: {prom_pronostico:.2f}")
    print("Tendencia:", "Al alza" if prom_pronostico > ultimo else "A la baja" if prom_pronostico < ultimo else "Estable")

# Procesar cada par
for ticker1, ticker2 in pairs:
    print(f"\n=== Análisis de {ticker1} y {ticker2} ===")

    # Cargar datos
    df1 = load_data(ticker1)
    df2 = load_data(ticker2)
    close1, close2 = prepare_data(df1, df2, ticker1, ticker2)

    # Pruebas de estacionariedad
    unit_root_tests(close1, ticker1)
    unit_root_tests(close2, ticker2)

    # Prueba de cointegración
    cointegration_test(close1, close2, ticker1, ticker2)

    # Ajustar modelos ARMA y pronosticar
    orden1 = find_best_arma(close1, ticker1)
    orden2 = find_best_arma(close2, ticker2)
    modelo1 = ARIMA(close1, order=orden1).fit()
    modelo2 = ARIMA(close2, order=orden2).fit()

    # Graficar pronósticos individuales
    plot_forecast(modelo1, close1, ticker1)
    plot_forecast(modelo2, close2, ticker2)

    # Graficar series históricas y pronósticos juntos
    pasos = 30
    pronostico1 = modelo1.forecast(steps=pasos)
    pronostico2 = modelo2.forecast(steps=pasos)
    indice_pronostico = range(len(close1), len(close1) + pasos)

    plt.figure(figsize=(12, 6))
    plt.plot(close1, label=f'Histórico {ticker1}')
    plt.plot(indice_pronostico, pronostico1, label=f'Pronóstico {ticker1}', color='red')
    plt.plot(close2, label=f'Histórico {ticker2}')
    plt.plot(indice_pronostico, pronostico2, label=f'Pronóstico {ticker2}', color='green')
    plt.title(f'Precios y pronósticos de {ticker1} y {ticker2}')
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import numpy as np
from google.colab import files
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.vector_ar.vecm import coint_johansen
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Lista de pares a analizar
pairs = [("XOM", "JNJ"), ("JNJ", "PG"), ("XOM", "PG")]

# Función para cargar datos
def load_data(ticker):
    print(f"Por favor, sube el archivo Excel para {ticker}")
    upload = files.upload()
    filename = list(upload.keys())[0]
    df = pd.read_excel(filename)
    return df

# Función para preparar datos
def prepare_data(df1, df2, ticker1, ticker2):
    print(f"\nColumnas del DataFrame {ticker1}:", list(df1.columns))
    print(f"Columnas del DataFrame {ticker2}:", list(df2.columns))

    columna_cierre = input(f"Ingresa el nombre de la columna con los precios de cierre para {ticker1} y {ticker2}: ")
    try:
        close1 = df1[columna_cierre]
        close2 = df2[columna_cierre]
    except KeyError:
        print(f"Error: La columna '{columna_cierre}' no se encuentra en uno o ambos DataFrames")
        raise

    min_longitud = min(len(close1), len(close2))
    return close1[:min_longitud].dropna(), close2[:min_longitud].dropna()

# Paso 1: Determinar tendencia y estacionalidad
def plot_trend_seasonality(series, name):
    plt.figure(figsize=(12, 6))
    plt.plot(series, label=f'Precios {name}')
    plt.title(f'Precios de cierre de {name}')
    plt.legend()
    plt.show()

    # Descomposición (período 30 como ejemplo, ajustar según datos)
    decomposition = seasonal_decompose(series, model='additive', period=30)
    tendencia = decomposition.trend
    estacionalidad = decomposition.seasonal
    residuo = decomposition.resid

    plt.figure(figsize=(12, 8))
    plt.subplot(411)
    plt.plot(series, label='Original')
    plt.legend(loc='upper left')
    plt.subplot(412)
    plt.plot(tendencia, label='Tendencia')
    plt.legend(loc='upper left')
    plt.subplot(413)
    plt.plot(estacionalidad, label='Estacionalidad')
    plt.legend(loc='upper left')
    plt.subplot(414)
    plt.plot(residuo, label='Residuos')
    plt.legend(loc='upper left')
    plt.tight_layout()
    plt.show()

    print(f"\nAnálisis para {name}:")
    print("- Tendencia: Observa el gráfico de tendencia - pendiente ascendente/descendente indica tendencia.")
    print("- Estacionalidad: Patrones repetitivos en el gráfico de estacionalidad sugieren estacionalidad.")

# Paso 2: Pruebas de raíz unitaria
def unit_root_tests(series, name):
    print(f"\nPruebas de raíz unitaria para {name}:")
    adf_result = adfuller(series)
    print("Prueba ADF:")
    print(f'Estadístico ADF: {adf_result[0]:.4f}, p-valor: {adf_result[1]:.4f}')
    print("Interpretación:", "Estacionario" if adf_result[1] < 0.05 else "No estacionario")

    kpss_result = kpss(series)
    print("\nPrueba KPSS:")
    print(f'Estadístico KPSS: {kpss_result[0]:.4f}, p-valor: {kpss_result[1]:.4f}')
    print("Interpretación:", "No estacionario" if kpss_result[1] < 0.05 else "Estacionario")

# Función para encontrar el mejor modelo ARMA
def find_best_arma(series, name, max_p=3, max_q=3):
    mejor_aic = float('inf')
    mejor_orden = None
    for p in range(max_p + 1):
        for q in range(max_q + 1):
            try:
                modelo = ARIMA(series, order=(p, 0, q))
                resultados = modelo.fit()
                if resultados.aic < mejor_aic:
                    mejor_aic = resultados.aic
                    mejor_orden = (p, 0, q)
            except:
                continue
    if mejor_orden is None:
        mejor_orden = (1, 0, 1)
    print(f"\nMejor modelo ARMA para {name}: {mejor_orden}, AIC: {mejor_aic:.2f}")
    return mejor_orden

# Graficar entrenamiento, prueba y pronóstico
def plot_train_test_forecast(entrenamiento, prueba, pronostico, name):
    indice_pronostico = range(len(entrenamiento), len(entrenamiento) + len(prueba))
    plt.figure(figsize=(12, 6))
    plt.plot(entrenamiento, label=f'{name} Entrenamiento')
    plt.plot(range(len(entrenamiento), len(entrenamiento) + len(prueba)), prueba, label=f'{name} Prueba')
    plt.plot(indice_pronostico, pronostico, label=f'{name} Pronóstico', color='red')
    plt.title(f'{name} Entrenamiento, Prueba y Pronóstico')
    plt.legend()
    plt.show()

# Procesar cada par
for ticker1, ticker2 in pairs:
    print(f"\n=== Análisis de {ticker1} y {ticker2} ===")

    # Cargar datos
    df1 = load_data(ticker1)
    df2 = load_data(ticker2)
    close1, close2 = prepare_data(df1, df2, ticker1, ticker2)

    # Paso 1: Tendencia y estacionalidad
    plot_trend_seasonality(close1, ticker1)
    plot_trend_seasonality(close2, ticker2)

    # Paso 2: Pruebas de estacionariedad
    unit_root_tests(close1, ticker1)
    unit_root_tests(close2, ticker2)

    # Paso 3: Dividir datos en entrenamiento (90%) y prueba (10%)
    tamano_entrenamiento = int(len(close1) * 0.9)
    train1, test1 = close1[:tamano_entrenamiento], close1[tamano_entrenamiento:]
    train2, test2 = close2[:tamano_entrenamiento], close2[tamano_entrenamiento:]

    print(f"\nDivisión de datos para {ticker1}: Entrenamiento={len(train1)}, Prueba={len(test1)}")
    print(f"División de datos para {ticker2}: Entrenamiento={len(train2)}, Prueba={len(test2)}")

    # Ajustar modelos ARMA
    orden1 = find_best_arma(train1, ticker1)
    orden2 = find_best_arma(train2, ticker2)
    modelo1 = ARIMA(train1, order=orden1).fit()
    modelo2 = ARIMA(train2, order=orden2).fit()

    # Pronosticar
    pasos_prueba = len(test1)
    pronostico1 = modelo1.forecast(steps=pasos_prueba)
    pronostico2 = modelo2.forecast(steps=pasos_prueba)

    # Graficar
    plot_train_test_forecast(train1, test1, pronostico1, ticker1)
    plot_train_test_forecast(train2, test2, pronostico2, ticker2)

    # Evaluar precisión
    mse1 = mean_squared_error(test1, pronostico1)
    mse2 = mean_squared_error(test2, pronostico2)
    print(f"\nPrecisión del pronóstico:")
    print(f"{ticker1} MSE: {mse1:.2f}")
    print(f"{ticker2} MSE: {mse2:.2f}")
```

```python
import pandas as pd
import numpy as np
from google.colab import files
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.vector_ar.vecm import coint_johansen
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Lista de pares a analizar
pairs = [("XOM", "JNJ"), ("JNJ", "PG"), ("XOM", "PG")]

# Variable para almacenar el texto consolidado
analisis_texto = ""

# Función para cargar datos
def load_data(ticker):
    print(f"Por favor, sube el archivo Excel para {ticker}")
    upload = files.upload()
    filename = list(upload.keys())[0]
    df = pd.read_excel(filename)
    return df

# Función para preparar datos
def prepare_data(df1, df2, ticker1, ticker2):
    print(f"\nColumnas del DataFrame {ticker1}:", list(df1.columns))
    print(f"Columnas del DataFrame {ticker2}:", list(df2.columns))
    columna_cierre = input(f"Ingresa la columna de precios de cierre para {ticker1} y {ticker2}: ")
    try:
        close1 = df1[columna_cierre]
        close2 = df2[columna_cierre]
    except KeyError:
        print(f"Error: La columna '{columna_cierre}' no se encuentra en uno o ambos DataFrames")
        raise
    min_longitud = min(len(close1), len(close2))
    return close1[:min_longitud].dropna(), close2[:min_longitud].dropna()

# Tendencia y estacionalidad
def analyze_trend_seasonality(series, name):
    decomposition = seasonal_decompose(series, model='additive', period=30)
    tendencia = decomposition.trend.dropna()
    estacionalidad = decomposition.seasonal.dropna()

    tendencia_desc = "ascendente" if tendencia.iloc[-1] > tendencia.iloc[0] else "descendente" if tendencia.iloc[-1] < tendencia.iloc[0] else "estable"
    estacionalidad_desc = "presente" if estacionalidad.std() > series.std() * 0.1 else "no significativa"

    return f"- Tendencia: {tendencia_desc}.\n- Estacionalidad: {estacionalidad_desc}."

# Pruebas de estacionariedad
def unit_root_tests(series, name):
    adf_result = adfuller(series)
    kpss_result = kpss(series)
    adf_inter = "Estacionario" if adf_result[1] < 0.05 else "No estacionario"
    kpss_inter = "No estacionario" if kpss_result[1] < 0.05 else "Estacionario"
    return (f"- ADF: Estadístico={adf_result[0]:.4f}, p-valor={adf_result[1]:.4f} ({adf_inter}).\n"
            f"- KPSS: Estadístico={kpss_result[0]:.4f}, p-valor={kpss_result[1]:.4f} ({kpss_inter}).")

# Cointegración
def cointegration_test(close1, close2, ticker1, ticker2):
    df = pd.DataFrame({ticker1: close1, ticker2: close2})
    result = coint_johansen(df, det_order=0, k_ar_diff=1)
    coint = "Sí" if result.lr1[0] > result.cvt[0, 1] else "No"
    return f"- Cointegración: {coint} (Estadístico de traza={result.lr1[0]:.2f}, Valor crítico 95%={result.cvt[0, 1]:.2f})."

# Modelo ARMA
def find_best_arma(series, name, max_p=3, max_q=3):
    mejor_aic = float('inf')
    mejor_orden = None
    for p in range(max_p + 1):
        for q in range(max_q + 1):
            try:
                modelo = ARIMA(series, order=(p, 0, q))
                resultados = modelo.fit()
                if resultados.aic < mejor_aic:
                    mejor_aic = resultados.aic
                    mejor_orden = (p, 0, q)
            except:
                continue
    if mejor_orden is None:
        mejor_orden = (1, 0, 1)
    return mejor_orden, mejor_aic

# Pronóstico y evaluación
def analyze_forecast(modelo, train, test, name):
    pasos = len(test)
    pronostico = modelo.forecast(steps=pasos)
    mse = mean_squared_error(test, pronostico)
    ultimo = train.iloc[-1]
    prom_pronostico = pronostico.mean()
    tendencia = "Al alza" if prom_pronostico > ultimo else "A la baja" if prom_pronostico < ultimo else "Estable"
    return (f"- Pronóstico: Último valor={ultimo:.2f}, Promedio pronosticado={prom_pronostico:.2f}.\n"
            f"- Tendencia: {tendencia}.\n- MSE: {mse:.2f}.")

# Procesar cada par y consolidar análisis
for ticker1, ticker2 in pairs:
    analisis_texto += f"\n=== Análisis de {ticker1} y {ticker2} ===\n"

    # Cargar y preparar datos
    df1 = load_data(ticker1)
    df2 = load_data(ticker2)
    close1, close2 = prepare_data(df1, df2, ticker1, ticker2)

    # Tendencia y estacionalidad
    analisis_texto += f"\nAnálisis de {ticker1}:\n"
    analisis_texto += analyze_trend_seasonality(close1, ticker1)
    analisis_texto += f"\nAnálisis de {ticker2}:\n"
    analisis_texto += analyze_trend_seasonality(close2, ticker2)

    # Estacionariedad
    analisis_texto += f"\nEstacionariedad de {ticker1}:\n"
    analisis_texto += unit_root_tests(close1, ticker1)
    analisis_texto += f"\nEstacionariedad de {ticker2}:\n"
    analisis_texto += unit_root_tests(close2, ticker2)

    # Cointegración
    analisis_texto += f"\nRelación entre {ticker1} y {ticker2}:\n"
    analisis_texto += cointegration_test(close1, close2, ticker1, ticker2)

    # Dividir datos
    tamano_entrenamiento = int(len(close1) * 0.9)
    train1, test1 = close1[:tamano_entrenamiento], close1[tamano_entrenamiento:]
    train2, test2 = close2[:tamano_entrenamiento], close2[tamano_entrenamiento:]

    # Modelos ARMA y pronósticos
    orden1, aic1 = find_best_arma(train1, ticker1)
    orden2, aic2 = find_best_arma(train2, ticker2)
    modelo1 = ARIMA(train1, order=orden1).fit()
    modelo2 = ARIMA(train2, order=orden2).fit()

    analisis_texto += f"\nModelo ARMA para {ticker1}: Orden={orden1}, AIC={aic1:.2f}\n"
    analisis_texto += analyze_forecast(modelo1, train1, test1, ticker1)
    analisis_texto += f"\nModelo ARMA para {ticker2}: Orden={orden2}, AIC={aic2:.2f}\n"
    analisis_texto += analyze_forecast(modelo2, train2, test2, ticker2)

    # Interpretación general
    analisis_texto += f"\nInterpretación general de {ticker1} y {ticker2}:\n"
    if "Sí" in cointegration_test(close1, close2, ticker1, ticker2):
        analisis_texto += "- Las series están cointegradas, sugiriendo una relación a largo plazo.\n"
    else:
        analisis_texto += "- No hay cointegración, las series se mueven independientemente a largo plazo.\n"
    analisis_texto += (f"- {ticker1}: Precisión del pronóstico depende de su estacionariedad y volatilidad "
                       f"(MSE alto indica menor ajuste).\n")
    analisis_texto += (f"- {ticker2}: Similarmente, un MSE bajo sugiere un modelo más confiable.\n")

# Imprimir el texto consolidado
print(analisis_texto)

# Opcional: Guardar en un archivo
with open("analisis_consolidado.txt", "w") as file:
    file.write(analisis_texto)
```

```python
import pandas as pd
import numpy as np
from google.colab import files
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Lista de tickers a analizar
tickers = ["XOM", "JNJ", "PG"]

# Variable para almacenar el texto consolidado
analisis_texto = "=== Análisis Individual y Comparación de Tendencias ===\n\n"

# Función para cargar datos
def load_data(ticker):
    print(f"Por favor, sube el archivo Excel para {ticker}")
    upload = files.upload()
    filename = list(upload.keys())[0]
    df = pd.read_excel(filename)
    return df

# Función para preparar datos
def prepare_data(df, ticker):
    print(f"\nColumnas del DataFrame {ticker}:", list(df.columns))
    columna_cierre = input(f"Ingresa la columna de precios de cierre para {ticker}: ")
    try:
        close = df[columna_cierre].dropna()
    except KeyError:
        print(f"Error: La columna '{columna_cierre}' no se encuentra en el DataFrame")
        raise
    return close

# Tendencia y estacionalidad
def analyze_trend_seasonality(series, name):
    decomposition = seasonal_decompose(series, model='additive', period=30)
    tendencia = decomposition.trend.dropna()
    estacionalidad = decomposition.seasonal.dropna()

    tendencia_desc = "ascendente" if tendencia.iloc[-1] > tendencia.iloc[0] else "descendente" if tendencia.iloc[-1] < tendencia.iloc[0] else "estable"
    estacionalidad_desc = "presente" if estacionalidad.std() > series.std() * 0.1 else "no significativa"

    return f"- Tendencia: {tendencia_desc}.\n- Estacionalidad: {estacionalidad_desc}."

# Pruebas de estacionariedad
def unit_root_tests(series, name):
    adf_result = adfuller(series)
    kpss_result = kpss(series)
    adf_inter = "Estacionario" if adf_result[1] < 0.05 else "No estacionario"
    kpss_inter = "No estacionario" if kpss_result[1] < 0.05 else "Estacionario"
    return (f"- ADF: Estadístico={adf_result[0]:.4f}, p-valor={adf_result[1]:.4f} ({adf_inter}).\n"
            f"- KPSS: Estadístico={kpss_result[0]:.4f}, p-valor={kpss_result[1]:.4f} ({kpss_inter}).")

# Modelo ARMA
def find_best_arma(series, name, max_p=3, max_q=3):
    mejor_aic = float('inf')
    mejor_orden = None
    for p in range(max_p + 1):
        for q in range(max_q + 1):
            try:
                modelo = ARIMA(series, order=(p, 0, q))
                resultados = modelo.fit()
                if resultados.aic < mejor_aic:
                    mejor_aic = resultados.aic
                    mejor_orden = (p, 0, q)
            except:
                continue
    if mejor_orden is None:
        mejor_orden = (1, 0, 1)
    return mejor_orden, mejor_aic

# Pronóstico y evaluación
def analyze_forecast(modelo, train, test, name):
    pasos = len(test)
    pronostico = modelo.forecast(steps=pasos)
    mse = mean_squared_error(test, pronostico)
    ultimo = train.iloc[-1]
    prom_pronostico = pronostico.mean()
    tendencia = "Al alza" if prom_pronostico > ultimo else "A la baja" if prom_pronostico < ultimo else "Estable"
    return (tendencia, mse, f"- Pronóstico: Último valor={ultimo:.2f}, Promedio pronosticado={prom_pronostico:.2f}.\n"
            f"- Tendencia: {tendencia}.\n- MSE: {mse:.2f}.")

# Diccionario para almacenar resultados de tendencias y MSE
resultados = {}

# Procesar cada ticker individualmente
for ticker in tickers:
    analisis_texto += f"Análisis de {ticker}\n"

    # Cargar y preparar datos
    df = load_data(ticker)
    close = prepare_data(df, ticker)

    # Tendencia y estacionalidad
    analisis_texto += "Tendencia y Estacionalidad:\n"
    analisis_texto += analyze_trend_seasonality(close, ticker)

    # Estacionariedad
    analisis_texto += "\nEstacionariedad:\n"
    analisis_texto += unit_root_tests(close, ticker)

    # Dividir datos
    tamano_entrenamiento = int(len(close) * 0.9)
    train, test = close[:tamano_entrenamiento], close[tamano_entrenamiento:]

    # Modelo ARMA y pronóstico
    orden, aic = find_best_arma(train, ticker)
    modelo = ARIMA(train, order=orden).fit()
    tendencia, mse, forecast_text = analyze_forecast(modelo, train, test, ticker)

    analisis_texto += f"\nModelo ARMA: Orden={orden}, AIC={aic:.2f}\n"
    analisis_texto += forecast_text

    # Almacenar resultados
    resultados[ticker] = {"tendencia": tendencia, "mse": mse}

    analisis_texto += "\n"

# Comparación y recomendación
analisis_texto += "=== Comparación y Recomendación de Inversión ===\n\n"
for ticker in tickers:
    analisis_texto += (f"{ticker}: Tendencia={resultados[ticker]['tendencia']}, "
                       f"MSE={resultados[ticker]['mse']:.2f}\n")

# Lógica de recomendación
tendencias = [resultados[ticker]["tendencia"] for ticker in tickers]
mses = {ticker: resultados[ticker]["mse"] for ticker in tickers}

if "Al alza" in tendencias:
    candidatos = [ticker for ticker in tickers if resultados[ticker]["tendencia"] == "Al alza"]
    mejor_candidato = min(candidatos, key=lambda x: mses[x])
    analisis_texto += (f"\nRecomendación: Invertir en {mejor_candidato}, ya que tiene una tendencia al alza "
                       f"con el menor MSE ({mses[mejor_candidato]:.2f}), indicando mayor precisión en el pronóstico.\n")
elif "Estable" in tendencias and "Al alza" not in tendencias:
    candidatos = [ticker for ticker in tickers if resultados[ticker]["tendencia"] == "Estable"]
    mejor_candidato = min(candidatos, key=lambda x: mses[x])
    analisis_texto += (f"\nRecomendación: Invertir en {mejor_candidato}, ya que es estable "
                       f"con el menor MSE ({mses[mejor_candidato]:.2f}), sugiriendo confiabilidad.\n")
else:
    analisis_texto += "\nRecomendación: No invertir en este momento, todas las tendencias son a la baja.\n"

# Imprimir el texto consolidado
print(analisis_texto)

# Guardar en un archivo
with open("analisis_inversion.txt", "w") as file:
    file.write(analisis_texto)
```
